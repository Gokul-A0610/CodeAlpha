```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate a random dataset
np.random.seed(42)
data = pd.DataFrame({
    'Age': np.random.randint(18, 65, 100),
    'Income': np.random.randint(20000, 100000, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Department': np.random.choice(['Sales', 'HR', 'Tech', 'Marketing'], 100)
})
```

```python
# Show first 5 rows
print("Sample Data:")
print(data.head())
```

```python
# Visualization 1:Distribution of age
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], kde=True, bins=15)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Distribution of Income
plt.figure(figsize=(8, 5))
sns.histplot(data['Income'], bins=15, kde=True, color='skyblue')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Visualization 3:Gender count
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
# Visualization 4:Department count
plt.figure(figsize=(8, 4))
sns.countplot(x='Department', data=data)
plt.title('Department Distribution')
plt.show()
```

```python
# Visualization 5:Income by gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='Income', data=data)
plt.title('Income Distribution by Gender')
plt.show()
```

```python
# Visualization 6: Age vs Income scatter plot with hue by Gender
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Income', hue='Gender', data=data, palette='Set2')
plt.title('Age vs Income by Gender')
plt.xlabel('Age')
plt.ylabel('Income')
plt.tight_layout()
plt.show()
```

```python
# Visualization 7:Income Distribution by Gender
plt.figure(figsize=(8, 5))
sns.violinplot(x='Gender', y='Income', data=data)
plt.title('Income Distribution by Gender')
plt.show()
```

```python
# Visualization 8:Income by Gender and Department
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Gender', y='Income', hue='Department', data=data)
plt.title('Income by Gender and Department')
plt.show()
```

```python
# Visualization 9:Pairplot of Age and Income Colored by Gender",
sns.pairplot(data, hue='Gender')
plt.suptitle("Pairplot of Age and Income Colored by Gender", y=1.02)
plt.show()
```

```python
# Visualization 10:Correlation Heatmap
plt.figure(figsize=(6, 4))
corr = data[['Age', 'Income']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualization 11:Average Income by Department
plt.figure(figsize=(8, 5))
sns.barplot(x='Department', y='Income', data=data, estimator=np.mean)
plt.title('Average Income by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 12:3D Scatter Plot: Age, Income, Department
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from sklearn.preprocessing import LabelEncoder

# Encode categorical values for 3D plotting
le_dept = LabelEncoder()
le_gender = LabelEncoder()
data['DeptEncoded'] = le_dept.fit_transform(data['Department'])
data['GenderEncoded'] = le_gender.fit_transform(data['Gender'])

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
sc = ax.scatter(
    data['Age'], 
    data['Income'], 
    data['DeptEncoded'],
    c=data['GenderEncoded'], 
    cmap=cm.coolwarm, 
    s=60
)

# Axis labels
ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_zlabel('Department')

# Add color bar legend
plt.colorbar(sc, ax=ax, label='Gender (0 = Female, 1 = Male)')

# Map encoded departments to tick labels
ax.set_zticks(range(len(le_dept.classes_)))
ax.set_zticklabels(le_dept.classes_)

plt.title('3D Scatter Plot: Age, Income, Department (Color = Gender)')
plt.show()
```
